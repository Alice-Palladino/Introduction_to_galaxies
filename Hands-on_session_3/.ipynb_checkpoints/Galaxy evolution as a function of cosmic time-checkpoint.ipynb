```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
```

### Extracting the data

```python
data = Table.read('3DHST_Allfields.fits')
```

```python
print(data)
```

```python
z = data['z_best']
z_err_l = data['z_best_l68']     # Lower error
z_err_u = data['z_best_u68']     # Upper error

umv = data['UmV']
vmj = data['VmJ']

mass = data['lmass']     # log10 of stellar mass
av = data['Av']     # dust extinction
lsfr = data['lsfr']     # log10 of star formation rate

ssfr = lsfr - mass     # Specific star formation rate, in log: log(sSFR) = log(SFR/mass) = log(SFR) - log(mass)
```

### Eliminating unphysical values
Masses too low and ssfr too low are unphysical. Let's eliminate them:

```python
# Specific star formation rate has a problem (listen to recordings for more info), we have to exclude the values 
# around -100 because they are not physical. Let's see, with an histogram, where the bulk of the data is:

fig = plt.figure(figsize=(10, 2))

for i in range(3):
    ax = fig.add_subplot(1,3,1+i)

    if i == 0:
        plt.hist(ssfr, bins = 30, color = 'mediumturquoise')
    if i == 1:
        plt.hist(ssfr, bins = np.linspace(-120, -30, 30), color = 'mediumturquoise')
    if i == 2:
        plt.hist(ssfr, bins = np.linspace(-30, 0, 30), color = 'mediumturquoise')
    
    plt.grid(c='grey', alpha = 0.5)

plt.tight_layout()
plt.show()

# PROF: take ssfr > -13: having a specific star formation rate of 10^-45 practically means zero.

ssfr_mask = ssfr > -13
print('Number of data still in array: %i' % len(ssfr[ssfr_mask]))
print('Number of excluded data: %i' % len(ssfr[~ssfr_mask]))
```

```python
fig = plt.figure(figsize=(10, 2))

for i in range(3):
    ax = fig.add_subplot(1,3,1+i)

    if i == 0:
        plt.hist(mass, bins = 30, color = 'mediumturquoise')
    if i == 1:
        plt.hist(mass, bins = np.linspace(-120, -60, 30), color = 'mediumturquoise')
    if i == 2:
        plt.hist(mass, bins = np.linspace(0, 20, 30), color = 'mediumturquoise')
    
    plt.grid(c='grey', alpha = 0.5)

plt.tight_layout()
plt.show()

# Prof: cut the masses from 8 

mass_mask = mass > 8
print('Number of data still in array: %i' % len(mass[mass_mask]))
print('Number of excluded data: %i' % len(mass[~mass_mask]))
```

```python
if np.all(~ssfr_mask[~mass_mask]):
    print("✅ All indices excluded by mass_mask are also excluded by ssfr_mask")
else:
    print("⚠️ Some indices excluded by mass_mask are NOT excluded by ssfr_mask")
```

```python
combined_mask = ssfr_mask & mass_mask

z = z[combined_mask]
z_err_l = z_err_l[combined_mask]
z_err_u = z_err_u[combined_mask]

umv = umv[combined_mask]
vmj = vmj[combined_mask]

mass = mass[combined_mask]
av = av[combined_mask]
lsfr = lsfr[combined_mask]
ssfr = ssfr[combined_mask]
```

```python
print('Minimum redshift: z = %.2f' % np.min(z))
print('Maximum redshift: z = %.2f' % np.max(z))
print('\n')

plt.hist(z, bins = 40, range=[np.min(z), np.max(z)], density = True, color = 'mediumturquoise')
plt.grid(alpha = 0.2, c = 'grey')
plt.xlabel('Redshift z')
plt.show()

############################################################################################################################
# "Cosa scelgo come intervalli?" Devi definire un subset di indici che rispettano certe condizioni -> mask oppure where.
# array[</> tot] -> true/false
# "Da scienziata, cosa decido per intervalli?" -> Matematicamente: "Ho tanti dati in certi intervalli e meno in altri -> 
# bin più stretti dove ho più dati, per maggiore precisione." Fisicamente: "Il redshift non è lineare, a redshift minori
# corrispondono tempi minori??? (Non mi è ben chiaro, però questa cosa ha detto qualcosa del genere del "la prima parte con
# un bin piccolo e la seconda parte con un bin largo e più o meno il tempo è lo stesso". Check!)
############################################################################################################################
```

```python
# Comincia da 0.5 perché prima hai quasi zero galassie

plt.hist(z, bins = [0.5, 0.9, 1.4, 2, np.max(z)], density = False, color = 'mediumturquoise')
plt.grid(alpha = 0.2, c = 'grey')
plt.xlabel('Redshift z')
plt.show()
```

```python
mask1 = (z>=0.5) & (z<0.9)
mask2 = (z>=0.9) & (z<1.4)
mask3 = (z>=1.4) & (z<2)
mask4 = z>=2

bin_edges = [0.5, 0.9, 1.4, 2]

z1 = z[mask1]
z2 = z[mask2]
z3 = z[mask3]
z4 = z[mask4]

print(len(z1))
print(len(z2))
print(len(z3))
print(len(z4))

print('\nLunghezze degli array')

print(len(z1) + len(z2) + len(z3) + len(z4))
print(len(z))

# Ne mancano tre prima del 0.5
```

```python
# Plotting, for all redshifts, the U-V vs V-J graph

plt.plot(vmj, umv, c = 'mediumturquoise', marker = 'o', ms = 3, ls='', alpha = 0.5)
plt.grid(alpha=0.2, c='grey')
plt.xlim([-1, 3])
plt.ylim([-0.5, 3])
plt.xlabel('U-V')
plt.ylabel('V-J')
plt.title('U-V vs V-J for all redshift')
plt.show()
```

```python
# Plotting the relation for each bin

fig = plt.figure(figsize=(10,7))

for i in range(4):
    ax = fig.add_subplot(2,2,1+i)
   
    if i == 0:
        mask = mask1
    if i == 1:
        mask = mask2
    if i == 2:
        mask = mask3
    if i == 3:
        mask = mask4
    
    plt.plot(vmj[mask], umv[mask], c = 'mediumturquoise', marker = 'o', ms = 3, ls='', alpha = 0.5)
    plt.grid(alpha=0.2, c='grey')
    plt.xlim([-1, 3])
    plt.ylim([-0.5, 3])
    plt.xlabel('U-V')
    plt.ylabel('V-J')
    if i < 3:
        plt.title('U-V vs V-J for %i° bin %.1f <= z < %.1f' % (i+1, bin_edges[i], bin_edges[i+1]))
    if i == 3:
        plt.title('U-V vs V-J for %i° bin z >= %.1f' % (i+1, bin_edges[i]))

plt.tight_layout()
plt.show()
```

```python
# Adding the dust attenuation

fig = plt.figure(figsize=(10,7))

for i in range(4):
    ax = fig.add_subplot(2,2,1+i)
   
    if i == 0:
        mask = mask1
    if i == 1:
        mask = mask2
    if i == 2:
        mask = mask3
    if i == 3:
        mask = mask4
    
    plt.scatter(vmj[mask], umv[mask], c = av[mask], cmap = 'rainbow', marker = 'o', s = 5, alpha = 0.2)
    plt.grid(alpha=0.2, c='grey')
    plt.xlim([-1, 3])
    plt.ylim([-0.5, 3])
    plt.xlabel('U-V')
    plt.ylabel('V-J')
    if i < 3:
        plt.title('U-V vs V-J for %i° bin %.1f <= z < %.1f' % (i+1, bin_edges[i], bin_edges[i+1]))
    if i == 3:
        plt.title('U-V vs V-J for %i° bin z >= %.1f' % (i+1, bin_edges[i]))
    plt.colorbar()

plt.tight_layout()
plt.show()
```
